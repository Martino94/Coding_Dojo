# K Nearest Neighbors


```python
import pandas as pd
import numpy as np
# Para escalar datos
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

El conjunto de datos Iris de scikit-learn no requiere descargar ningún archivo de algún sitio web externo. El código de abajo carga el conjunto de datos Iris.

```python
data = load_iris()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

Organizar los datos en una matriz de características y un vector de destino

```python
X = df.loc[:, ['sepal length (cm)', 
               'sepal width (cm)',
               'petal length (cm)',
               'petal width (cm)']]
```

```python
# La matriz debe ser bidimensional
X.shape
```

```python
y = df.loc[:, ['target']].values
```

```python
y.shape
```

La estandarización de un conjunto de datos es un requisito común para muchos estimadores de aprendizaje automático

```python
scaler = StandardScaler()
# Ajusta en X (asumiendo que tu matriz de características es una matriz NumPy)
scaler.fit(X)
# Transform X
X = scaler.transform(X)
```

```python
X
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# Mirando algunos hiperparámetros que parecen ajustables
print(knn)
```

```python
knn.fit(X, y)
```

```python
predictions = knn.predict(X)
```

```python
predictions
```

```python
score = knn.score(X, y)
score
```

### Si KNN puede llegar al 100% de precisión o un R2 de 1, ¿por qué no se usa con más frecuencia?

La precisión fue de 1.0 o 100%. Este modelo parecía funcionar muy, muy bien. Ajustar nuestro modelo a los mismos datos utilizados para hacer predicciones (y encontrar precisión) no es una práctica recomendada. El aprendizaje automático consiste en crear modelos que predigan el futuro y no el pasado. Se trata de hacer modelos que se generalicen bien a nuevos datos. En módulos futuros, revisaremos la validación del modelo, lo que ayudará a que un modelo se generalice bien a nuevos datos. KNN, en particular, es excelente para predecir datos que ya tienes y no tanto para predecir el futuro. Este modelo puede obtener un 100% de precisión en la mayoría de los casos. Hay excepciones, como si los datos en los que se ajustaron tienen valores de características duplicados y algunos valores de destino diferentes.
