```python
from matplotlib import pyplot
```

```python
x = df['Percentage of Dog Owners']
y = df['Percentage of Cat Owners']
```

```python
plt.style.use('seaborn')
pyplot.hist(x, alpha=0.5, bins=20, label='Percentage of Dog Owners')
pyplot.hist(y, alpha=0.5, bins=20, label='Percentage of Cat Owners')
pyplot.legend(loc='upper right')
pyplot.show()
```

La distribución del porcentaje de dueños de perros está acumulada hacia la derecha. La distribución del porcentaje de dueños de gatos está acumulada
en el centro. No obstante, por la falta de datos no se puede ver el comportamiento real de las variables.


```python
plt.boxplot([x,y], notch = True, labels=['Percentage of Dog Owners', 'Percentage of Cat Owners'],  patch_artist=True,
           medianprops = dict(linestyle='-', linewidth=2, color='Yellow'),
           boxprops = dict(linestyle='--', linewidth=2, color='Black', facecolor = 'blue', alpha = .4)
          )

plt.title('Diagrama de cajas')
```

En promedio, hay más personas que son duenñas de perros, no obstante, en dos estados hay datos atípicos en el extremo superior sobre el porcentaje de dueños de gatos.
